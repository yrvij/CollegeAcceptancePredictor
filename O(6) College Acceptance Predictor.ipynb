```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
import random

possibilities = ['accepted','rejected']
X = []
Y = []

for i in range(5000):
    X.append([random.randint(1300,1601),round(random.uniform(3.5,5.0),1),random.randint(1,20)])
    Y.append(random.choice(possibilities))

P = [[1467, 4.2, 2]]

#{DecisionTreeClassifier}
counter = 0
clf = DecisionTreeClassifier()
clf = clf.fit(X,Y)
print ("1) Using Decision Tree Prediction is " + clf.predict(P)[0])
if clf.predict(P) == 'accepted':
    counter += 1
    
#{KNeighborsClassifier}
knn = KNeighborsClassifier()
knn.fit(X,Y)
print ("2) Using K Neighbors Classifier Prediction is " + knn.predict(P)[0])
if knn.predict(P) == 'accepted':
    counter += 1
    
#{MLPClassifier}
mlpc = MLPClassifier()
mlpc.fit(X,Y)
print ("3) Using MLPC Classifier Prediction is " + mlpc.predict(P)[0])
if mlpc.predict(P) == 'accepted':
    counter += 1

#{RandomForestClassifier}
rfor = RandomForestClassifier()
rfor.fit(X,Y)
print ("4) Using RandomForestClassifier Prediction is " + rfor.predict(P)[0])
if rfor.predict(P) == 'accepted':
    counter += 1
    
#{LogisticRegression}
logr = LogisticRegression()
logr.fit(X,Y)
print("5) Using LogisticRegression Prediction is " + logr.predict(P)[0])
if logr.predict(P) == 'accepted':
    counter += 1

#{NaiveBayesClassifier}
gnb = GaussianNB()
gnb.fit(X,Y)
print("6) Using NaiveBayesClassifier Prediction is " + gnb.predict(P)[0])
if gnb.predict(P) == 'accepted':
    counter += 1

print()
print(P)
print()
print('You have approximately a ' + str(round(((counter/6)*100),2)) + '% chance of acceptance.')
```
